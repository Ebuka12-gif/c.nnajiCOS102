```python
import tkinter as tk
from tkinter import ttk

class BankDivision:
    def __init__(self, services):
        self.services = services

    def get_services(self):
        return self.services

class RetailBanking(BankDivision):
    def __init__(self):
        services = [
            "Lines of credit",
            "Investment management and accounts",
            "Insurance",
            "Retirement and education accounts",
            "Loans and mortgages",
            "Checking and saving"
        ]
        super().__init__(services)

class GlobalBanking(BankDivision):
    def __init__(self):
        services = [
            "Multi-currency management services and products",
            "Foreign currency accounts",
            "Foreign currency credit cards",
            "Transborder advisory services",
            "Liquidity management"
        ]
        super().__init__(services)

class CommercialBanking(BankDivision):
    def __init__(self):
        services = [
            "Lines of credit",
            "Investment management and accounts",
            "Insurance",
            "Advisory services"
        ]
        super().__init__(services)

class BankApp:
    def __init__(self, root):
        self.root = root
        self.root.title("Zenith Bank Nigeria Services")

        self.name_label = tk.Label(root, text="Employee Name:")
        self.name_label.grid(row=0, column=0, padx=10, pady=10)

        self.name_entry = tk.Entry(root)
        self.name_entry.grid(row=0, column=1, padx=10, pady=10)

        self.division_label = tk.Label(root, text="Division:")
        self.division_label.grid(row=1, column=0, padx=10, pady=10)

        self.division_combobox = ttk.Combobox(root, values=["Retail Banking", "Global Banking", "Commercial Banking"])
        self.division_combobox.grid(row=1, column=1, padx=10, pady=10)

        self.display_button = tk.Button(root, text="Display Services", command=self.display_services)
        self.display_button.grid(row=2, columnspan=2, padx=10, pady=10)

        self.services_text = tk.Text(root, width=50, height=10)
        self.services_text.grid(row=3, columnspan=2, padx=10, pady=10)

    def display_services(self):
        name = self.name_entry.get()
        division = self.division_combobox.get()
        self.services_text.delete(1.0, tk.END)

        if division == "Retail Banking":
            division_obj = RetailBanking()
        elif division == "Global Banking":
            division_obj = GlobalBanking()
        elif division == "Commercial Banking":
            division_obj = CommercialBanking()
        else:
            self.services_text.insert(tk.END, "Invalid division selected")
            return

        services = division_obj.get_services()
        services_str = "\n".join(services)

        self.services_text.insert(tk.END, f"Employee Name: {name}\n")
        self.services_text.insert(tk.END, f"Division: {division}\n")
        self.services_text.insert(tk.END, "Services Provided:\n")
        self.services_text.insert(tk.END, services_str)

if __name__ == "__main__":
    root = tk.Tk()
    app = BankApp(root)
    root.mainloop()

```
