```python
import tkinter as tk
from tkinter import messagebox

order_entry = None
quantity_entry = None
total_price = 0  # Initialize total price

def button_click():
    # Show an information message box 
    messagebox.showinfo("Info", "Welcome to the PAU cafeteria!.\n")

    # Ask for user confirmation
    result = messagebox.askyesno("Confirmation", "Do you want to make an order?")

# Create the main window
root = tk.Tk()
root.title("Home Page")
root.geometry("600x500")

items = ["jollof rice", "coconut fried rice", "jollof spaghetti", "sweet chili chicken", "grilled chicken wings", "fried beef", "fried fish", "boiled egg", "sauteed sausages", "savoury beans", "roasted sweet potatoes", "fried plantains", "mixed vegetable salad", "boiled yam", "eba", "poundo yam", "sermo", "alama soup", "egusi soup", "water", "glass drink[55cl]", "pet drink[35cl]", "pet drink[50cl]", "glass/canned malt", "fresh yo", "pineapple juice", "mango juice", "zubo drink"]
prices = [350, 350, 350, 1100, 400, 400, 500, 200, 200, 350, 300, 150, 150, 150, 100, 100, 100, 450, 480, 200, 150, 300, 500, 500, 600, 350, 350, 350]

def menu():
    mainmenu_label = tk.Label(root, text = "RICE/PASTA          PRICE($)                      PROTEINS              PRICE($)\n[1]Jollof Rice:        350                        [4]Sweet Chili Chicken:  1100\n[2]Cocunut Fired Rice: 350                        [5]Grilled Chicken Wings: 400\n[3]Jollof Spaghetti:   350                        [6]Fried Beef:            400\n                                                  [7]Fried Fish:            500\n                                                  [8]Boiled Egg:            200\n                                                  [9]Sauteed Sausages:      200\n\n")
    mainmenu_label.pack()
    sidedish_label = tk.Label(root, text = "                           SIDE DISHES              PRICE($)\n                           [10]Savoury beans:           350\n                           [11]Roasted Sweet Potatoes:  300\n                           [12]Fried Plantains:         150\n                           [13]Mixed Vegetable Salad:   150\n                           [14]Boiled Yam:              150\n\n")
    sidedish_label.pack()
    swallowdrinks_label = tk.Label(root, text = "SOUPS & SWALLOW        PRICE($)             BEVERAGES            PRICE($)\n[15]Eba:                   100               [20]Water:               200\n[16]Poundo Yam:            100               [21]Glass Drink[55cl]:   150\n[17]Sermo:                 100               [22]PET Drink[35cl]:     300\n[18]Alama Soup:            450               [23]PET Drink[50cl]:     500\n[19]Egusi Soup:            480               [24]Glass/Canned Malt:   500\n                                             [25]Fresh Yo:            600\n                                             [26]Pineapple Juice:     350\n                                             [27]Mango Juice:         350\n                                             [28]Zubo Drink:          350")
    swallowdrinks_label.pack()

menu()  # Call the menu function to display the menu first

def submit_order():
    global total_price
    
    # Extract order details from entry widget
    order_index = int(order_entry.get()) - 1
    quantity = int(quantity_entry.get())

    # Calculate total price for the item
    item_price = prices[order_index]
    item_total_price = item_price * quantity
    
    # Add item price to total price
    total_price += item_total_price

    # Ask the user if they want to buy any other item
    result = messagebox.askyesno("Confirmation", "Do you want to buy any other item?")

    if result:
        # Clear entry fields for the next order
        order_entry.delete(0, tk.END)
        quantity_entry.delete(0, tk.END)
    else:
        # Apply discount based on total price
        discount = 0
        if total_price < 1000:
            discount = 0
        elif total_price <= 2500:
            discount = total_price * 0.1
        elif total_price < 5000:
            discount = total_price * 0.15
        else:
            discount = total_price * 0.25

        # Subtract discount from total price
        total_price -= discount

        # Display order confirmation message
        messagebox.showinfo("Order Confirmation", f"Your order has been placed.\nTotal price after discount: ${total_price}")

        # Clear entry fields
        order_entry.delete(0, tk.END)
        quantity_entry.delete(0, tk.END)

# Create widgets for order entry
order_label = tk.Label(root, text="Enter the item number you want to order:")
order_label.pack()

order_entry = tk.Entry(root)
order_entry.pack()

quantity_label = tk.Label(root, text="Enter the quantity:")
quantity_label.pack()

quantity_entry = tk.Entry(root)
quantity_entry.pack()

submit_button = tk.Button(root, text="Submit Order", command=submit_order)
submit_button.pack()

root.mainloop()
```
